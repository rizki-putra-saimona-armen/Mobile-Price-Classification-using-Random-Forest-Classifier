```python
df = pd.read_csv("data/mobile.csv")
df.head()
```

```python
plot_missing_value(df)
```

# dataset Spliting

```python
X = df.drop(columns="price_range")
y = df.price_range

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y,  random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from jcopml.tuning import grid_search_params as gsp
```

```python
X_train.head()
```

```python
gsp.rf_params
```

```python
preprocessor = ColumnTransformer([
    ('numeric', num_pipe(), [
        "battery_power", "clock_speed", "fc", "int_memory", "m_dep", "mobile_wt",
        "n_cores", "pc", "px_height", "px_width", "ram", "sc_h", "sc_w", "talk_time"
    ]),
    ('categoric', cat_pipe(encoder='onehot'), [
        "blue", "dual_sim", "four_g", "three_g", "touch_screen", "wifi"
    ]),
])

pipeline = Pipeline([
    ("prep", preprocessor),
    ("algo",RandomForestClassifier(n_jobs=-1, random_state=42))
])

model = GridSearchCV(pipeline, gsp.rf_params, cv=3, n_jobs =-1, verbose=1)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

# Feature Importance

```python
df_imp = mean_score_decrease(X_train, y_train, model, plot=True, topk=10)
```

# Phase 2

```python
preprocessor = ColumnTransformer([
    ('numeric', num_pipe(), ["battery_power", "mobile_wt", "px_height", "px_width","ram"]),
])

pipeline = Pipeline([
    ("prep", preprocessor),
    ("algo",RandomForestClassifier(n_jobs=-1, random_state=42))
])

model = GridSearchCV(pipeline, gsp.rf_params, cv=3, n_jobs =-1, verbose=1)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

# Phase 3: Try Polynomial after feture Selection

```python
preprocessor = ColumnTransformer([
    ('numeric', num_pipe(poly = 2), ["battery_power", "mobile_wt", "px_height", "px_width","ram"]),
])

pipeline = Pipeline([
    ("prep", preprocessor),
    ("algo",RandomForestClassifier(n_jobs=-1, random_state=42))
])

model = GridSearchCV(pipeline, gsp.rf_params, cv=3, n_jobs =-1, verbose=1)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

# RandomizedSearchCV

Karena GridSearch mencoba semua kombinasi yang kita tentukan, prosesnya bisa memakan waktu yang sangat lama. Oleh karena itu, GridSearch juga dikenal sebagai Exhaustive Search. Pendekatan yang lebih efisien adalah dengan tidak mencoba semua kombinasi, melainkan memilih sejumlah kombinasi tertentu secara acak. Metode pencarian semacam ini disebut RandomizedSearch.

Keunggulan dari RandomizedSearch adalah kita tidak harus menentukan angka pasti, tetapi dapat memberikan distribusi pencarian yang fleksibel. Namun, kelemahannya adalah metode ini tidak menjamin kita mendapatkan parameter yang optimal karena hanya mengevaluasi sebagian dari keseluruhan kombinasi yang tersedia.

```python
from sklearn.model_selection import RandomizedSearchCV
from jcopml.tuning import random_search_params as rsp
```

```python
rsp.rf_poly_params
```

```python
preprocessor = ColumnTransformer([
    ('numeric', num_pipe(poly=2), ["battery_power", "mobile_wt", "px_height", "px_width","ram"]),
])

pipeline = Pipeline([
    ("prep", preprocessor),
    ("algo",RandomForestClassifier(n_jobs=-1, random_state=42))
])

model = RandomizedSearchCV(pipeline, rsp.rf_poly_params, cv=3, n_iter=50, n_jobs =-1, verbose=1, random_state=42)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```
